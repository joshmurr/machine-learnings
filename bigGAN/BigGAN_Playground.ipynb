````
# Import some Python packages...
import torch
import torchvision
import nltk
nltk.download('wordnet')

import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
from google.colab import files

from pytorch_pretrained_biggan import (BigGAN,
                                       BigGANConfig,
                                       one_hot_from_names, 
                                       truncated_noise_sample,
                                       save_as_images, 
                                       display_in_terminal)


# Choose the SIZE of the model you want to use here.
# Bigger means larger images and more detail, but it runs a lot slower.
# I recommend you get to grips with the model on 'biggan-deep-256' and
# then if you want to make some higher quality images move onto 'biggan-deep-512'

model = BigGAN.from_pretrained('biggan-deep-128')
#model = BigGAN.from_pretrained('biggan-deep-256')
# model = BigGAN.from_pretrained('biggan-deep-512')
````

````
# These are a bunch of helper functions...
def llerp(A, B, num_interps):
  if A.shape != B.shape:
    raise ValueError('A and B must have the same shape to interpolate.')
  alphas = np.linspace(0, 1, num_interps)
  return np.array([(1-a)*A + a*B for a in alphas])
  
def slerp_wiki(v0, v1, interval):
    """Spherical linear interpolation."""
    # https://en.wikipedia.org/wiki/Slerp
    # >>> slerp([1,0,0,0], [0,0,0,1], np.arange(0, 1, 0.001))
    steps = np.arange(0, 1, 1/interval)
    t_array = np.array(steps)
    v0 = np.array(v0)
    v1 = np.array(v1)
    dot = np.sum(v0 * v1)

    if dot < 0.0:
        v1 = -v1
        dot = -dot
    
    DOT_THRESHOLD = 0.9995
    if dot > DOT_THRESHOLD:
        result = v0[np.newaxis,:] + t_array[:,np.newaxis] * (v1 - v0)[np.newaxis,:]
        return (result.T / np.linalg.norm(result, axis=1)).T
    
    theta_0 = np.arccos(dot)
    sin_theta_0 = np.sin(theta_0)

    theta = theta_0 * t_array
    sin_theta = np.sin(theta)
    
    s0 = np.cos(theta) - dot * sin_theta / sin_theta_0
    s1 = sin_theta / sin_theta_0
    return (s0[:,np.newaxis] * v0[np.newaxis,:]) + (s1[:,np.newaxis] * v1[np.newaxis,:])
  
def show_image(_tensor):
  img = _tensor.permute(1, 2,0).detach().numpy()
  img = np.clip(((img + 1) / 2.0) * 256, 0, 255)
  plt.imshow(np.asarray(np.uint8(img), dtype=np.uint8))

def save_image(imageName, _tensor):
  img = get_image(_tensor, permute=[2, 1, 0])
  cv2.imwrite(imageName, img)
  files.download(imageName)

def make_video(video_name, imgs, codec='DIVX'):
  _, _, height, width = imgs.shape
  video = cv2.VideoWriter(video_name, cv2.VideoWriter_fourcc(*codec), fps=24, frameSize=(width,height))
  for iter in range(0,imgs.shape[0]):
    frame = get_image(imgs[iter], permute=[2,1,0])
    video.write(frame)
  cv2.destroyAllWindows()
  video.release()
  files.download(video_name)

def get_image(_tensor, permute=[0, 1, 2]):
  if len(_tensor.size()) == 4:
    img = _tensor[0]
  else:
    img = _tensor
  img = img.permute(1, 2, 0).detach().numpy()
  img = img[:, :, permute]
  img = np.clip(((img + 1) / 2.0) * 256, 0, 255)
  return np.asarray(np.uint8(img), dtype=np.uint8)

FOLDER = 0
def save_images(_tensor, dir_name=None, file_name='frame'):
  global FOLDER
  dir_name = dir_name or f'{FOLDER:03}'
  path = os.path.join('/content', dir_name)
  os.mkdir(path)
  os.chdir(dir_name)
  save_as_images(_tensor, file_name=file_name)
  os.chdir('/content/')
  FOLDER = FOLDER + 1

  print(f'Images saved to /{dir_name}')
````

````
# Some more functions...

def run_model(class_tensor, noise_tensor, truncation):
  noise_vector = torch.from_numpy(noise_tensor).float()
  class_vector = torch.from_numpy(class_tensor).float()

  type(noise_vector)

  noise_vector = noise_vector.to('cuda')
  class_vector = class_vector.to('cuda')
  model.to('cuda')

  with torch.no_grad():
      output = model(noise_vector, class_vector, truncation)

  return output.to('cpu')

def show_grid(tensor):
  grid = torchvision.utils.make_grid(tensor, nrow=12)
  plt.figure(figsize = (12, int(tensor.shape[0] / 12)))
  grid = grid.permute(1, 2,0).detach().numpy()
  grid = np.clip(((grid + 1) / 2.0) * 256, 0, 255)
  plt.imshow(np.asarray(np.uint8(grid), dtype=np.uint8))

def generate_image(classLabels, scale=None, noiseVector=None, truncation=0.4):
  scale = scale or [1] * len(classLabels)

  class_vector = one_hot_from_names(classLabels, batch_size=len(classLabels))
  noise_vector = truncated_noise_sample(truncation=truncation, batch_size=1)

  class_vector = [cv * s for cv, s in zip(class_vector, scale)]
  class_vector = sum(class_vector)
  class_vector = np.expand_dims(class_vector, axis=0)

  output = run_model(class_vector, noise_vector, truncation)
  show_image(output[0])

  return output

def generate_thumbnail_loop(
    classLabels=['dog'],
    scale=None,
    truncation=0.7,
    numFrames=60,
    lerpFunc=llerp, # Linear Lerp seems to work best
    wave=None,
    noise=None):
  scale = scale or [1] * len(classLabels)

  if len(classLabels) < 2:
    classLabels *= 2

  class_vector = one_hot_from_names(classLabels, batch_size=len(classLabels))
  noise_vector = truncated_noise_sample(truncation=truncation, batch_size=len(classLabels))

  i=0
  interval = numFrames // len(classLabels)
  class_lerps = lerpFunc(class_vector[i], class_vector[(i+1)%len(classLabels)], interval)
  noise_lerps = lerpFunc(noise_vector[i], noise_vector[(i+1)%len(classLabels)], interval)
  i += 1

  while i < len(classLabels):
    class_lerps = np.concatenate((class_lerps, lerpFunc(class_vector[i], class_vector[(i+1)%len(classLabels)], interval)))
    noise_lerps = np.concatenate((noise_lerps, lerpFunc(noise_vector[i], noise_vector[(i+1)%len(classLabels)], interval)))
    i += 1

  if wave:
    # Non-periodic sine wave
    noise_lerps = np.sin(wave * noise_lerps) + np.sin(np.pi * noise_lerps)

  if noise:
    noise_signal = np.random.normal(0, noise, noise_lerps.shape)
    noise_lerps += noise_signal

  output = run_model(class_lerps, noise_lerps, truncation)
  display_grid(output)

  return output
 
def generate_thumbnail_pair(classLabels=['dog'], previous=None, scale=None, truncation=0.4, numFrames=12):
  scale = scale or [1] * len(classLabels)

  if len(classLabels) == 1:
    classLabels *= 2

  class_vector = []
  noise_vector = []

  if previous:
    class_vector.append(previous['class_vector'])
    noise_vector.append(previous['noise_vector'])
    class_vector.append(one_hot_from_names(classLabels[0], batch_size=1)[0])
    noise_vector.append(truncated_noise_sample(truncation=truncation, batch_size=1)[0])
  else:
    class_vector = one_hot_from_names(classLabels, batch_size=len(classLabels))
    noise_vector = truncated_noise_sample(truncation=truncation, batch_size=len(classLabels))

  class_lerps = llerp(class_vector[0], class_vector[1], numFrames)
  noise_lerps = llerp(noise_vector[0], noise_vector[1], numFrames)

  next = { 'class_vector': class_lerps[-1], 'noise_vector': noise_lerps[-1] }

  output = run_model(class_lerps, noise_lerps, truncation)
  display_grid(output)

  return output, next
````

# __Generating Single Images__

OK finally let's make some images. We will use the function `generate_image()` to make a single image. Huggingface have helpfully integrated some natural language processing so you can try any word and it will attempt to find the closest match, but it wont always work. See the list of words __at the end of the page__ to be more specific.

The function takes the form:

```python
output = generate_image(classLabels=['dog'], truncation=0.4)
```

Where `classLabels` are a list of the labels you're searching for and `truncation` is a special variable which controls how wildly around the area specified to search. The truncation factor refers to the breadth of the latent space to sample. The values close to zero are the most frequently sampled in training so tend to return the best results now, however we don't always want the most realistic looking image! So values closer to 1 will return more unexpected results.

Try changing the values in the function below!

````
swan = generate_image(classLabels=['swan'], truncation=0.2)
````

# __Mix Two Classes__

We can add more class labels to the list and with some simple arithmetic we can combine them. It's literally as simple as adding two more more numbers together. However to keep things in a manageable range we __scale__ the numbers slightly, hence `scale=[0.7, 0.7]` in the call to `generate_image()`. The numbers in `scale` will scale the respective label in `classLabels` by that amount. So if we want slightly more tennis ball than dog we would pass `scale=[0.7, 0.8]` into the function.

It's possible to pass more than two classes into this function, have a play.

````
dog_tennis = generate_image(classLabels=['dog', 'tennis_ball'], truncation=0.6, scale=[0.7, 0.8])
````

````
save_image('dog_tennis.png', dog_tennis)
````

# __Generate a Sequence of Images__

Producing a sequence of images is a much more interesting way to get a sense of the latent space of BigGAN. `generate_thumbnail_loop()` is the function we will use to do just that. 

By adding multiple class labels it will interpolate from one point to the next and will automatically create a loop for you.

````
dog_elephant = generate_thumbnail_loop(classLabels=['dog', 'elephant'], numFrames=120, truncation=0.7)
````

# __Turn the Sequence Into a Video!__

Obviously we want to see the sequence as a video so we have the function `save_video()` to do just that. The function does not return anything so you don't need to assign it to a variable and will automatically start the downloading for you.

Pass the name of your sequence (above) into the function like so:

````
# Turn a thumbnail sequence into a video!

#               Name of sequence from generate_thumbnails()
#                                   |
#         Name of video             |
#              |                    |
#              |                    |
save_video('dog_elephant.mp4', dog_elephant) 
````

Sometimes the results can be a little bit boring. To add some variety there are two ways you can add some variation to the path we are taking through the latent space as we interpolate from point to point.

Firstly you can add an argument called `wave` into our function. This adds a non-periodic (it doesn't repeat) wave to the path add some variety. `wave=1` is a small amount of variation and `wave=100` is a high-frequency wave and will behave randomly.

````
dogs = generate_thumbnail_loop(classLabels=['dog', 'elephant'], numFrames=120, truncation=0.7, wave=2)
````

Another option is to literally add noise to the numbers. This naturally is uncontrollable but can be interesting to just get a selection of random images.

Any value above 0.3 will add a lot of randomness!

````
dog_elephant_random = generate_thumbnail_loop(classLabels=['dog', 'elephant'], numFrames=120, truncation=0.7, noise=0.5)
````

# __Colab Constraints__

Colab will only give you a certain amount of memory and playing with BigGAN can very quickly use up your allowance, especially if you are using the 512x512 model size.

If this happens you will see an error like so:
```
RuntimeError: CUDA out of memory. Tried to allocate 2.00 MiB (GPU 0; 15.90 GiB total capacity; 15.18 GiB already allocated; 1.88 MiB free; 15.19 GiB reserved in total by PyTorch)
```

To get around this you will need to restart your runtime to clear the memory by going to `Runtime -> Restart Runtime`. You may need to re-run the cells at the start of the notebook as we are effectively starting again.

To counter this we can use `generate_thumbnail_pair()` to make a shorter sequence of images. This function will return the latent vector which the sequence ends on. In the example below I am saving that vector in `next` and then passing that into the function again (but seperately) to start the next sequence where the previous left off! That way we can ask BigGAN to do smaller tasks at a time and not saturate our memory allowance, whilst still being able to control how me move through latent space.

Don't forget to save the images or video as you go! Save as soon as you see something you like.

````
# Choose the first two class labels.
turtle_to_spider, next = generate_thumbnail_pair(classLabels=['leatherback turtle', 'harvestman'], numFrames=12, truncation=0.7)
````

````
# Choose the next class label and pass `next` from before, into the following function
spider_to_truck, next = generate_thumbnail_pair(classLabels=['truck'], previous=next, numFrames=12, truncation=0.7)
````

````
# And again... as many times as you wish.
truck_to_elephant, next = generate_thumbnail_pair(classLabels=['elephant'], previous=next, numFrames=24, truncation=0.7)
````
